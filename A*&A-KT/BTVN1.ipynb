```python
# -*- coding: utf-8 -*-
import heapq
import random

class NPuzzleAKT:
    def __init__(self, n, initial_state, goal_state):
        self.n = n
        self.initial_state = initial_state
        self.goal_state = goal_state
        self.adj_moves = [(1, 0), (0, -1), (-1, 0), (0, 1)]

    def h(self, state):
        """Hàm Heuristic Manhattan Distance"""
        dist = 0
        goal_map = {self.goal_state[i][j]: (i, j) for i in range(self.n) for j in range(self.n)}
        for i in range(self.n):
            for j in range(self.n):
                val = state[i][j]
                if val != 0:
                    target_i, target_j = goal_map[val]
                    dist += abs(i - target_i) + abs(j - target_j)
        return dist

    def get_neighbors(self, state):
        neighbors = []
        r, c = next((i, j) for i in range(self.n) for j in range(self.n) if state[i][j] == 0)
        for dr, dc in self.adj_moves:
            nr, nc = r + dr, c + dc
            if 0 <= nr < self.n and 0 <= nc < self.n:
                new_state = [list(row) for row in state]
                new_state[r][c], new_state[nr][nc] = new_state[nr][nc], new_state[r][c]
                neighbors.append(tuple(map(tuple, new_state)))
        return neighbors

    def solve(self):
        start_node = tuple(map(tuple, self.initial_state))
        goal_node = tuple(map(tuple, self.goal_state))

        open_lst = [(self.h(self.initial_state), 0, self.h(self.initial_state), start_node)]
        closed_lst = set()
        g = {start_node: 0}
        par = {start_node: start_node}

        print(f"\n--- Đang tìm đường đi ngắn nhất bằng AKT ---")

        while open_lst:
            f_n, g_n, h_n, n_state = heapq.heappop(open_lst)

            if n_state == goal_node:
                self.print_result(par, n_state, g[n_state])
                return

            closed_lst.add(n_state)
            for m_state in self.get_neighbors(n_state):
                new_g = g_n + 1
                if m_state not in closed_lst:
                    if m_state not in g or new_g < g[m_state]:
                        g[m_state] = new_g
                        par[m_state] = n_state
                        h_m = self.h(m_state)
                        heapq.heappush(open_lst, (new_g + h_m, new_g, h_m, m_state))

    def print_result(self, par, current, total_g):
        path = []
        while par[current] != current:
            path.append(current)
            current = par[current]
        path.append(current)
        path.reverse()

        print(f"\n[THÀNH CÔNG] Đã tìm thấy lời giải tối ưu!")
        print(f"Tổng số bước di chuyển cần thiết: {total_g}")
        print("-" * 35)
        for idx, s in enumerate(path):
            print(f"Bước {idx}:")
            for row in s:
                print("  " + " ".join(f"{x:2}" for x in row))
        print("==> HOÀN THÀNH!")

def generate_puzzle(n, shuffle_steps=20):
    # Tạo đích chuẩn
    goal = [[(i * n + j + 1) % (n * n) for j in range(n)] for i in range(n)]
    state = [row[:] for row in goal]

    # Xáo trộn ngược từ đích để đảm bảo giải được
    for _ in range(shuffle_steps):
        r, c = next((i, j) for i in range(n) for j in range(n) if state[i][j] == 0)
        moves = [(r+dr, c+dc) for dr, dc in [(1,0), (-1,0), (0,1), (0,-1)]
                 if 0 <= r+dr < n and 0 <= c+dc < n]
        nr, nc = random.choice(moves)
        state[r][c], state[nr][nc] = state[nr][nc], state[r][c]
    return state, goal

def print_matrix(matrix, title):
    print(f"\n[MA TRẬN {title.upper()}]")
    for row in matrix:
        print("  " + " ".join(f"{x:2}" for x in row))

# --- CHƯƠNG TRÌNH CHÍNH ---
if __name__ == "__main__":
    try:
        n_input = int(input("Nhập kích thước n (vd: 3, 4, 5): "))

        # Sinh tự động
        initial, goal = generate_puzzle(n_input)

        # In ra cho người dùng thấy
        print_matrix(initial, "đầu (Khởi tạo)")
        print_matrix(goal, "đích (Mục tiêu)")

        print("\n" + "="*35)
        input("Nhấn Enter để bắt đầu giải...")

        solver = NPuzzleAKT(n_input, initial, goal)
        solver.solve()

    except ValueError:
        print("Lỗi: Vui lòng chỉ nhập số nguyên!")
    except Exception as e:
        print(f"Lỗi hệ thống: {e}")
```
