```python
# -*- coding: utf-8 -*-
from heapq import heappush, heappop

class PriorityQueue:
    def __init__(self):
        self.heap = []
    def push(self, key):
        heappush(self.heap, key)
    def pop(self):
        return heappop(self.heap)
    def empty(self):
        return not self.heap

class Node:
    def __init__(self, name, parent=None, g=0, h=0):
        self.name = name
        self.parent = parent
        self.g = g  # Chi phí thực tế từ điểm bắt đầu
        self.h = h  # Chi phí ước lượng đến đích (Heuristic)
        self.f = g + h # Tổng chi phí dự kiến

    # So sánh f(n) để PriorityQueue ưu tiên node tốt nhất
    def __lt__(self, nxt):
        return self.f < nxt.f

def a_star_search(adj_list, h_table, start, goal):
    # Danh sách Mở (Open List) dùng Priority Queue để tối ưu
    open_list = PriorityQueue()
    # Danh sách Đóng (Closed List) để lưu các đỉnh đã duyệt xong
    closed_list = set()

    # Khởi tạo đỉnh bắt đầu
    start_node = Node(start, None, 0, h_table.get(start, 0))
    open_list.push(start_node)

    # Lưu g_cost thấp nhất từng đạt được của mỗi đỉnh
    best_g = {start: 0}

    step = 0
    print(f"--- BẮT ĐẦU TÌM ĐƯỜNG TỪ {start} ĐẾN {goal} ---")

    while not open_list.empty():
        step += 1
        # Lấy node có f(n) nhỏ nhất
        current_node = open_list.pop()
        u = current_node.name

        # In thông tin chi tiết bước duyệt
        print(f"Bước {step}: Xét đỉnh '{u}'")
        print(f"  -> g={current_node.g}, h={current_node.h}, f={current_node.f}")

        # Kiểm tra nếu đã tới đích
        if u == goal:
            path = []
            while current_node:
                path.append(current_node.name)
                current_node = current_node.parent
            path.reverse()
            print("\n==> ĐÃ TÌM THẤY ĐƯỜNG ĐI!")
            print(f"Lộ trình: {' -> '.join(path)}")
            print(f"Tổng chi phí thực tế: {best_g[goal]}")
            return path

        closed_list.add(u)

        # Duyệt các đỉnh kề (neighbors)
        for neighbor, weight in adj_list.get(u, []):
            if neighbor in closed_list:
                continue

            new_g = current_node.g + weight

            # Nếu tìm thấy đường đi tới neighbor tốt hơn đường cũ
            if neighbor not in best_g or new_g < best_g[neighbor]:
                best_g[neighbor] = new_g
                h_val = h_table.get(neighbor, 0)
                neighbor_node = Node(neighbor, current_node, new_g, h_val)
                open_list.push(neighbor_node)

        print(f"  - Closed List: {list(closed_list)}")
        print("-" * 35)

    print("Không tìm thấy đường đi!")
    return None

# --- DỮ LIỆU ĐẦU VÀO ---
# Danh sách kề: Đỉnh -> [(Đỉnh kề, Trọng số cạnh), ...]
adjac_lis = {
    'A': [('B', 2), ('C', 3)],
    'B': [('D', 3), ('E', 1)],
    'C': [('E', 1)],
    'D': [('F', 2)],
    'E': [('F', 4)],
    'F': []
}

# Bảng Heuristic: Khoảng cách ước lượng từ mỗi đỉnh đến F
h_table = {
    'A': 6,
    'B': 4,
    'C': 4,
    'D': 2,
    'E': 3,
    'F': 0
}

# Thực thi tìm đường từ A đến F
a_star_search(adjac_lis, h_table, 'A', 'F')
```
