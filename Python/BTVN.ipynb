```python
def tron_mang(a, b):
    c = []
    n = min(len(a), len(b))

    for i in range(n):
        c.append(a[i] + b[i])

    if len(a) > len(b):
        c.extend(a[n:])
    else:
        c.extend(b[n:])

    return c
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

kq = tron_mang(a, b)
print("Mảng a: ", a)
print("Mảng b: ", b)
print("Kết quả: ", kq)
```

```python
#a. Tổng tam giác trên của chéo phụ
def tong_tam_giac_tren(a):
    n = len(a)
    tong = 0

    for i in range(n):
        for j in range(n):
            if i + j < n - 1:
                tong+=a[i][j]

    return tong

#b. Doi so am thanh tri tuyet doi
def doi_dau(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])

#c. Thay các phần tử chẵn trong a bằng số nguyên
def thay_phan_tu(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] == x

#d. Kiểm tra a có toàn chẵn
def toan_chan(a):
  for row in a:
    for x in row:
      if x % 2 != 0:
        return False
  return True

# e. Kiểm tra đối xứng
def doi_xung(a):
  n = len(a)
  for i in range(n):
    for j in range(n):
      if a[i][j] != a[j][i]:
        return False
  return True

# f. Đường chéo chính tăng dần
def cheo_chinh_tang(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][i] > a[i + 1][i + 1]:
            return False
    return True


# g. Xuất tam giác dưới chéo phụ (kể cả chéo phụ)
def xuat_tam_giac_duoi_cheo_phu(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:  # dưới + chéo phụ
                print(a[i][j], end=" ")
        print()


# h. Chéo phụ giảm dần
def cheo_phu_giam(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][n - 1 - i] < a[i + 1][n - 2 - i]:
            return False
    return True


# ===== TEST =====
a = [
    [1, -2, 3],
    [4, 5, -6],
    [7, 8, 9]
]

print("Mảng a: ", a)

print("Tổng tam giác trên chéo phụ:", tong_tam_giac_tren(a))

doi_dau(a)
print("Sau đổi âm:", a)

thay_phan_tu(a, 0)
print("Sau thay chẵn:", a)

print("Toàn chẵn:", toan_chan(a))
print("Đối xứng:", doi_xung(a))
print("Chéo chính tăng:", cheo_chinh_tang(a))

print("Tam giác dưới chéo phụ:")
xuat_tam_giac_duoi_cheo_phu(a)

print("Chéo phụ giảm:", cheo_phu_giam(a))
```
